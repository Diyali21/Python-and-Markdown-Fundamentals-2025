```python
print(np.median(marks))
```

### Task 1
At the every year tell me balance for 10 years
Clue: General formula
1102.5 -> 1000 +
[1050, 1102.5 ...., ]

```python
INTEREST_RATE = 0.05
amount = 1_000

amt_list = []

for i in range(10):
    amount *= (1 + INTEREST_RATE)
    amt_list.append(round(amount, 2))

print(amt_list)

year_10 = round(amt_list[9] * 1.05, 2)
print(year_10)
```

```python
investment = 1_000
interest_rate = 1.05
num_years = np.arange(1, 11)

print(investment * interest_rate ** num_years)

year_10_interest = investment * interest_rate ** num_years[-1]

print(year_10_interest)
```
