# Task 1

# Task 2.2

````python
tips_df.groupby('smoker').aggregate({'tip': 'mean'})

# To display sum and mean
tips_df.groupby('smoker').aggregate({'tip': ['mean', 'sum']})
````

````python
# Also include total_bill avg

tips_df.groupby("smoker").aggregate({"tip": ["mean", "sum"], 'total_bill': 'mean'})
````

Transforms the result pulling out 'group by column' in separate columns

````python
tips_df.groupby(["day", "sex"])["tip"].mean().unstack()
````

## Visualize

- Matplot
  - Pioneer
  - Beginner-friendly

- Seaborn

````python
import matplotlib.pyplot as plt
import seaborn as sns
````

### Bar Graph

````python
import numpy as np

y_axis = tips_df['tip'].head(5)
x_axis = y_axis.index + 1

# x_axis = np.arange(5, 10)

plt.bar(x_axis, y_axis, color="plum", alpha = 0.8)
plt.xlabel('Days')
plt.ylabel('Tips')

plt.title('Tips on the first 5 days')
````

## Task
 
Draw a bar graph for Average total Bill for each Day with nice labels

````python
y_axis = tips_df.groupby("day")["total_bill"].mean()
x_axis = y_axis.index

plt.bar(x_axis, y_axis, color = 'purple', alpha=0.5)
plt.xlabel('Days')
plt.ylabel('Average Total Bill')
plt.title('Average Total Bill per Day')
````

````python
y_avg_total_bill = tips_df.groupby("day")["total_bill"].mean()

y_avg_total_bill.plot(
    kind='bar',
    xlabel='Days',
    ylabel='Average sales in R',
    color = 'purple',
    alpha = 0.8)
````

````python
y_avg_total_bill = tips_df.groupby("day")["total_bill"].mean()

y_avg_total_bill.plot(
    kind="pie",
    autopct = '%1.2f%%',
    title='Average sales in the week'
)
````

## Task: Proportion of Meals
Create a pie chart comparing how many times people visited on which time of the day (Dinner, Lunch)

````python
y_axis = tips_df['time'].value_counts()

y_axis.plot(
    kind = 'pie',
    autopct = '%1.1f%%',
    title = 'Proportion of Meals'
)
````

````python
tips_df["tips_percent"] = round((tips_df["tip"] / tips_df["total_bill"]) * 100, 2)
plt.scatter(tips_df["total_bill"], tips_df["tips_percent"], color="plum")

plt.xlabel("Total Bill  (R)")
plt.ylabel("Tip percent %")
plt.title("Tip percent vs Total Bill")
````

````python
plt.hist(tips_df['total_bill'], color='orange', bins=15)
plt.xlabel("Total Bill (R)")
plt.ylabel('Frequency')
````

# Seaborn

````python
sns.barplot(x='day', y='total_bill', data=tips_df, palette='pastel')
plt.xlabel('Days')
plt.ylabel('Average sales in R')
plt.title('Average sales in the week')
plt.xticks(rotation=45)
````

````python
sns.regplot(x='total_bill', y='tip', data=tips_df, color='green')
plt.title('Total Bill vs. Tip with Regression Line')
plt.xlabel('Total Bill (R)')
plt.ylabel('Tip (R)')
````
